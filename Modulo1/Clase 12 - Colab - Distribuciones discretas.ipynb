## **Tipos de Distribuciones Discretas**

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
```

### Distribución binomial

```python
N, p = 30, 0.4 # parametros de forma 
binomial = stats.binom(N, p) # Distribución
x = np.arange(binomial.ppf(0.01),
              binomial.ppf(0.99))
fmp = binomial.pmf(x) # Función de Masa de Probabilidad
plt.plot(x, fmp, '--')
plt.vlines(x, 0, fmp, colors='g', lw=5, alpha=0.5)
plt.title('Distribución Binomial')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

### Distribución Hipergeométrica

```python
k, n, N = 30, 10, 12 # parametros de forma 
hipergeometrica = stats.hypergeom(k, n, N) # Distribución
x = np.arange(0, n+1)
fmp = hipergeometrica.pmf(x) # Función de Masa de Probabilidad
plt.plot(x, fmp, '--')
plt.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
plt.title('Distribución Hipergeométrica')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

¿Como se ve el ejercicio que desarrollamos en la explicación?

### Distribución Poisson

```python

mu =  3.4# parametro de forma 
poisson = stats.poisson(mu) # Distribución
x = np.arange(poisson.ppf(0.01),
              poisson.ppf(0.99))
fmp = poisson.pmf(x) # Función de Masa de Probabilidad
plt.plot(x, fmp, '--')
plt.vlines(x, 0, fmp, colors='r', lw=5, alpha=0.5)
plt.title('Distribución Poisson')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

Tambien definimos una función para calcular combinatoria:

```python
# Genere el código para calcular la combinatoria de dos números n y p
def combinatoria(n, p):
    factorial_n = math.factorial(n)
    factorial_pn = math.factorial(n-p)
    factorial_p = math.factorial(p)

    Combinatoria = factorial_n/(factorial_p*factorial_pn)

    return Combinatoria
```

```python
n = 3
p = 2
print(combinatoria(n,p))
```
