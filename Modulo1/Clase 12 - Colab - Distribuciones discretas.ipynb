## **Tipos de Distribuciones Discretas**

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
```

### Distribuci贸n binomial

```python
N, p = 30, 0.4 # parametros de forma 
binomial = stats.binom(N, p) # Distribuci贸n
x = np.arange(binomial.ppf(0.01),
              binomial.ppf(0.99))
fmp = binomial.pmf(x) # Funci贸n de Masa de Probabilidad
plt.plot(x, fmp, '--')
plt.vlines(x, 0, fmp, colors='g', lw=5, alpha=0.5)
plt.title('Distribuci贸n Binomial')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python
k, n, N = 30, 10, 12 # parametros de forma 
hipergeometrica = stats.hypergeom(k, n, N) # Distribuci贸n
x = np.arange(0, n+1)
fmp = hipergeometrica.pmf(x) # Funci贸n de Masa de Probabilidad
plt.plot(x, fmp, '--')
plt.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
plt.title('Distribuci贸n Hipergeom茅trica')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

Tambien definimos una funci贸n para calcular combinatoria:

```python
# Genere el c贸digo para calcular la combinatoria de dos n煤meros n y p
def combinatoria(n, p):
    factorial_n = math.factorial(n)
    factorial_pn = math.factorial(n-p)
    factorial_p = math.factorial(p)

    Combinatoria = factorial_n/(factorial_p*factorial_pn)

    return Combinatoria
```

```python
n = 3
p = 2
print(combinatoria(n,p))
```
